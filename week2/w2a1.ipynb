```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

### **1)** Generate a fake data set with exactly 2 features and at least 1000 points and  using the `sklearn.datasets.make_blobs` or `sklearn.datasets.make_circles`.

Using make_blobs I set the sample size to 1000 and the features to 2. I am using 24 as my random state so it produces the same result each time. It won't be a binary dataset by default, to get that we would need to set `centers=2`.
I put the `X`'s into a dataframe and change the `y` into a numpy array.

```python
X, y = make_blobs(n_samples=1000, n_features=2, random_state=24)
fake_df = pd.DataFrame(X, columns=['x1', 'x2'])
npy = y
```

### **2)** Create a scatterplot of the data. Based on this scatterplot, would you expect Logistic Regression to perform well in predicting the classes?

I don't know how good Logistic Regression is good for non-binary classification, but the groups are easily separable on the scatterplot, so I think Logistic Regression will work well on it.

```python
sns.scatterplot(data=fake_df, x='x1', y='x2', hue=npy)
```

### **3)** Create X and y dataframes based on the fake data set.

Did this in question 1.

### **4)** Call `sklearn.model_selection.train_test_split` to create `X_train`, `X_test`, `y_train`, and `y_test` data. Use a `test_size` of 0.3. 

This one is pretty straight forward. Plugging in `fake_df` which holds my `X`'s and npy which holds my `y`'s into `train_test_split()`.

```python
X_train, X_test, y_train, y_test = train_test_split(fake_df, npy, test_size=.3, random_state=24)
```

### **5)** Create a `LogisticRegression` object and call `.fit` to train the `LogisticRegression` object. 

This one is also pretty straight forward. Storing the `LogisticRegression` in `fake_fit`.

```python
fake_fit = LogisticRegression(random_state=24).fit(X_train, y_train)
```

### **6)** Call the `.score` method on your `LogisticRegression` object on your test data. How is the performance? 

It got 100 percent accuracy.

```python
fake_fit.score(X_test, y_test)
```
